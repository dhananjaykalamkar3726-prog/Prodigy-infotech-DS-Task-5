## Handling Missing Files

```python
#dropping columns which has more than 2500 missing values and Time column
df.drop(['Service_year_of_vehicle','Defect_of_vehicle','Work_of_casuality','Fitness_of_casuality','Time'],
        axis = 1,inplace =True)
df.head()
```

```python
#storing catgorical columns names to a new variable 
categorical=[i for i in df.columns if df[i].dtype=='O']
print('The categorical variables are',categorical)
```

## Handling Categorical Values

```python
df.dtypes
```

```python
# importing Label encoding module
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# creating a new data frame for performing the chi2 analysis
df1 = pd.DataFrame()

# adding all the categorical columns except the output to new data frame
for i in categorical:
    if i != 'Accident_severity':
        df1[i] = le.fit_transform(df[i])
```

```python
#confirming the data type
df1.info()
```

```python
plt.figure(figsize=(22,17))
sns.set(font_scale=1)
sns.heatmap(df1.corr(),annot=True)
```

```python
#label encoded data set
df1.head()
```

```python
#import chi2 test
from sklearn.feature_selection import chi2
f_p_values=chi2(df1,df['Accident_severity'])
```

```python
#f_p_values will return fscore and pvalues
f_p_values
```

```python
#for better understanding and ease of access them to a new dataframe
f_p_values1 = pd.DataFrame({'features': df1.columns, 'Fscore': f_p_values[0], 'Pvalues': f_p_values[1]})
f_p_values1   
```

```python
#since we want lower Pvalues we are sorting the features
f_p_values1.sort_values(by='Pvalues',ascending=True)
```

```python
#after evaluating we are removing lesser important columns and storing to a new data frame
df2=df.drop(['Owner_of_vehicle','Type_of_vehicle','Road_surface_conditions','Pedestrian_movement',
    'Casualty_severity','Educational_level','Day_of_week','Sex_of_driver','Road_allignment',
    'Sex_of_casualty'],axis=1)
df2.head()
```

```python
#get_dummies
dummy=pd.get_dummies(df2[['Age_band_of_driver','Vehicle_driver_relation','Driving_experience','Area_accident_occured','Lanes_or_Medians','Types_of_Junction','Road_surface_type','Light_conditions','Weather_conditions','Type_of_collision','Vehicle_movement','Casualty_class','Age_band_of_casualty','Cause_of_accident']],drop_first=True)
dummy.head()
```

```python
#concatinate dummy and old data frame
df3=pd.concat([df2,dummy],axis=1)
df3.head()
```

```python
#dropping dummied columns
df3.drop(['Age_band_of_driver','Vehicle_driver_relation','Driving_experience','Area_accident_occured','Lanes_or_Medians','Types_of_Junction','Road_surface_type','Light_conditions','Weather_conditions','Type_of_collision','Vehicle_movement','Casualty_class','Age_band_of_casualty','Cause_of_accident'],axis=1,inplace=True)
df3.head()
```

## Seperation Independent and Dependent

```python
x=df3.drop(['Accident_severity'],axis=1)
x.shape
```

```python
x.head()
```

```python
y=df3.iloc[:,2]
y.head()
```

```python
#checking the count of each item in the output column
y.value_counts()
```

```python
#plotting count plot using seaborn
sns.countplot(x = y,palette='muted')
```

## Oversampling

```python
#importing smote
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo=oversample.fit_resample(x,y)
```

```python
#checking the oversampling output
y1=pd.DataFrame(yo)
y1.value_counts()
```

```python
sns.countplot(x=yo,palette='muted')
```

## Splitting the Data

```python
#converting data and testing data 
from sklearn.model_selection import train_test_split
#splitting 70% of the data t training data and 30% of daa to testing data
x_train,x_test,y_train,y_test=train_test_split(xo,yo,test_size=0.30,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

## KNN Model Creation

## Prediction

```python
#KNN model alg
from sklearn.neighbors import KNeighborsClassifier
model_KNN=KNeighborsClassifier(n_neighbors=5)
model_KNN.fit(x_train,y_train)
```

```python
y_pred=model_KNN.predict(x_test)
```

```python
y_pred
```

## Checking Accuracy , Classification Report < Confusion Matrix

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,ConfusionMatrixDisplay
```

```python
report_KNN=classification_report(y_test,y_pred)
print(report_KNN)
```

```python
accuracy_KNN=accuracy_score(y_test,y_pred)
print(accuracy_KNN)
```

```python
matrix_KNN=confusion_matrix(y_test,y_pred)
print(matrix_KNN,'\n')
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
